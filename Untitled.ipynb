# Data Analysis 

```python


# Get the value counts for 'country_or_area' column
country_counts = df['country_or_area'].value_counts()

# Display the counts
print(country_counts)
```

```python
# Get the value counts for 'country_or_area' column
category_counts = df['category'].value_counts()

# Display the counts
print(category_counts)
```

```python
# Summary statistics for 'value' by 'country_or_area'
country_stats = df.groupby('country_or_area')['value'].describe()
print(country_stats)

# Summary statistics for 'value' by 'year'
year_stats = df.groupby('year')['value'].describe()
print(year_stats)
```

# Time Series Analysis

```python
import matplotlib.pyplot as plt

# Group by year and sum the values
yearly_data = df.groupby('year')['value'].sum().reset_index()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(yearly_data['year'], yearly_data['value'], marker='o')
plt.title('Total Greenhouse Gas Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('Emissions (in metric tons)')
plt.grid(True)
plt.show()
```

# Country-wise Analysis

```python
# Group by country and sum the values
country_data = df.groupby('country_or_area')['value'].sum().reset_index()

# Sort by emissions
country_data = country_data.sort_values(by='value', ascending=False)

# Plot the top 10 countries by emissions
plt.figure(figsize=(12, 6))
plt.bar(country_data['country_or_area'][:10], country_data['value'][:10])
plt.title('Top 10 Countries by Greenhouse Gas Emissions')
plt.xlabel('Country')
plt.ylabel('Emissions (in metric tons)')
plt.xticks(rotation=45)
plt.show()
```

# Category-wise Analysis

```python
# Group by category and sum the values
category_data = df.groupby('category')['value'].sum().reset_index()

# Sort by emissions
category_data = category_data.sort_values(by='value', ascending=False)

# Plot the data
plt.figure(figsize=(12, 6))
plt.bar(category_data['category'], category_data['value'])
plt.title('Greenhouse Gas Emissions by Category')
plt.xlabel('Category')
plt.ylabel('Emissions (in metric tons)')
plt.xticks(rotation=90)
plt.show()
```

# Trend Analysis by Country

```python
# List of selected countries
selected_countries = ['European Union', 'United States of America', 'Japan', 'Austria']

plt.figure(figsize=(14, 8))
for country in selected_countries:
    # Strip any extra whitespace from the country name
    country = country.strip()
    # Filter and group the data by year for the current country
    country_data = df[df['country_or_area'] == country].groupby('year')['value'].sum().reset_index()
    # Plot the emissions trend
    plt.plot(country_data['year'], country_data['value'], marker='o', label=country)

plt.title('Greenhouse Gas Emissions Trends by Country')
plt.xlabel('Year')
plt.ylabel('Emissions (in metric tons)')
plt.legend()
plt.grid(True)
plt.show()
```

# Year-on-Year Growth Rate

```python
# Calculate year-on-year growth rate
yearly_data['growth_rate'] = yearly_data['value'].pct_change() * 100

# Plot the growth rate
plt.figure(figsize=(12, 6))
plt.bar(yearly_data['year'], yearly_data['growth_rate'])
plt.title('Year-on-Year Growth Rate of Greenhouse Gas Emissions')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.show()
```

# Correlation Analysis

```python
# Pivot the data for correlation analysis
pivot_df = df.pivot_table(index='year', columns='category', values='value', aggfunc='sum').fillna(0)

# Compute correlation matrix
correlation_matrix = pivot_df.corr()

# Plot the correlation matrix
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Greenhouse Gas Emissions by Category')
plt.show()
```

# Seasonal Decomposition of Time Series

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series for a selected country
selected_country = 'Australia'
country_data = df[df['country_or_area'] == selected_country].groupby('year')['value'].sum().reset_index()

# Perform seasonal decomposition
decomposition = seasonal_decompose(country_data['value'], model='additive', period=1)

# Plot the decomposition
decomposition.plot()
plt.suptitle(f'Seasonal Decomposition of {selected_country} Greenhouse Gas Emissions')
plt.show()
```

# Predictive Modeling

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit an Exponential Smoothing model
model = ExponentialSmoothing(yearly_data['value'], trend='add', seasonal=None, seasonal_periods=1)
fit = model.fit()

# Forecast future emissions
forecast = fit.forecast(steps=10)  # Forecasting for the next 10 years

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(yearly_data['year'], yearly_data['value'], label='Observed')
plt.plot(range(yearly_data['year'].max()+1, yearly_data['year'].max()+11), forecast, label='Forecast', linestyle='--')
plt.title('Greenhouse Gas Emissions Forecast')
plt.xlabel('Year')
plt.ylabel('Emissions (in metric tons)')
plt.legend()
plt.grid(True)
plt.show()
```
